```python
x=[0,100,200,300,400,500,600,700,800,900,1000]
y=[125,125,120,112,90,90,95,88,75,35,0]

def simpsoncompuesto(x,y):
    par=0
    impar=0
    h = (x[-1]-x[0])/len(x)
    for i in range(len(x)):
      if(x[i] % 2 == 0):
        par = par + y[i]
      else:
        impar = impar + y[i]
    return (h/3)*(y[0]+y[-1]+4*impar+2*par)

simpsoncompuesto(x,y)
```
