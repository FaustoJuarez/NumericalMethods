```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3]
y = [1.3,1.5,1.8,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(-2, 15.5, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
img = plt.imread("duck.png")
ax = plt.axes()
ax.imshow(img,extent=[-2.8,18,-2,2.8])
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Interpolacion Cubic Spline')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

#Intente que coincida con la imagen pero no logro ajustar los parametros
```
